```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import gamma
from scipy.integrate import quad
```

```python
def f(x):
    return np.sin(gamma(x+1))/gamma(x+1)
    #return x**2
```

```python
xs = np.linspace(0,10,10000)
fig,ax = plt.subplots(figsize=(10,7))
plt.plot(xs,f(xs))
plt.ylabel(r'$\frac{sin(x!)}{x!}$',fontsize=20)
plt.xlabel(r'$x$',fontsize=20)
plt.xlim(0,5)
plt.grid()
plt.savefig('integrand.png',dpi=300)
```

```python
grid = np.linspace(0,100,1000)
res = []
for g in grid:
    res.append(quad(f,0,g)[0])
```

```python
phi = (1+np.sqrt(5))/2

fig,ax = plt.subplots(ncols=2,figsize=(16,7))

ax[0].scatter(grid,res,s=5)
ax[0].set_ylim(1.6,1.62)
ax[0].axhline(phi,c='r',label='$\phi$')
ax[0].legend(fontsize=20)
ax[0].set_xlabel("Z",fontsize=20)
ax[0].set_ylabel(r'$\int_0^{z} \frac{sin(x!)}{x!} dx$',fontsize=20)

ax[1].set_ylabel(r'$|\int_0^{z} \frac{sin(x!)}{x!} dx - \phi|$',fontsize=20)
diff = np.abs(res-phi)
ax[1].plot(grid,diff)
ax[1].set_yscale('log')
ax[1].set_xlabel("Z",fontsize=20)
plt.savefig('integral.png',dpi=300)
```
